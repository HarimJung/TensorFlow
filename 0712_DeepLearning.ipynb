# NAND Gate

```
import numpy as np
def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5]) # 부호만 바꿔서 위치를 reverse 할수있다.
    b=0.7 # bias 역시 부호 반대
    tmp=np.sum(x*w)+b
    if tmp<=0: # 반대의 개념이라 =를 포함.
      return 0
    else:
      return 1
```

```
NAND(1,1)
```

# OR Gate

```
def OR(x1,x2):
  x=np.array([x1,x2])
  w=np.array([3,4])
  b=-2
  
  tmp=np.sum(x*w)+b
  if tmp >=0:
    return 1
  else:
    return 0
```

```
OR(0,0)
```

```
OR(1,1)
```

```
OR(0,1)
```

```
OR(1,0)
```

```
for i in range(0,2):
  for j in range(0,2):
    result = OR(i,j)
    print("OR %d,%d is %d" %(i,j,result))
```

# AND Gate

```
def AND(x1,x2):
  x=np.array([x1,x2])
  w=np.array([-0.5,-0.5])
  b=0.7
  tmp=np.sum(x*w)+b
  if tmp<=0:
    return 0
  else:
    return 1
    
```

# Gigantic Linear Regression

```
#학습데이터(x,y)
x_train=[1,2,3,4]
y_train=[6,7,8,9]

#변수선언

w=tf.Variable(tf.random_normal([1]),name="weight") # 하나의 값을 뽑아내라, 보통 씨드값을 넣어준다.
b=tf.Variable(tf.random_normal([1]),name="bias")

# 가설정의, 랜덤하게 잡아서 넣어준다
# set the hypothesis, put the x_train randomly
hypo=x_train*w +b

#put the cost and lost function
# cost= get the minimum (hypothesis - y_train square) 
cost=tf.reduce_mean(tf.square(hypo-y_train)) # 하이포에서 y빼서 제곱해서, 평균 최소값을 내라.
```

```
import tensorflow as tf
import matplotlib.pyplot as plt
x=[1,2,3]
y=[1,2,3]

w=tf.placeholder(tf.float32)
hypo=x*w


cost= tf.reduce_mean(tf.square(hypo-y))

sess=tf.Session()

sess.run(tf.global_variables_initializer())

w_val=[]
cost_val=[]

for i in range(-30,50):
  feed_w=i*0.1
  cur_cost,cur_w=sess.run([cost,w],feed_dict={w:feed_w})# w 에 feed_w 값을넣어줌.
  w_val.append(cur_w)
  cost_val.append(cur_cost)
  
plt.plot(w_val,cost_val)
plt.show()
```

```
x_train=[1,2,3,4]
y_train=[6,5,7,10]

w=tf.Variable(tf.random_normal([1]),name="weight")
b=tf.Variable(tf.random_normal([1]),name="bias")

hypo=w*x_train+b

cost=tf.reduce_mean(tf.square(hypo-y_train))

opti=tf.train.GradientDescentOptimizer(learning_rate=0.01)

train=opti.minimize(cost)

sess=tf.Session()

sess.run(tf.global_variables_initializer())


for step in range(2000):
  sess.run(train)
  print(step,sess.run(cost),sess.run(w),sess.run(b))
  

plt.plot(w,cost)
plt.show()
```
